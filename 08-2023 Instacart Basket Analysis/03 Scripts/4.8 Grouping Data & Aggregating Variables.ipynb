```python
#Creating 'max order' column

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#checking results of previous code

ords_prods_merge.head(15)
```

```python
#creating 'loyal customer' flag

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#creating 'Regular customer' flag

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#creating 'New customer' flag

ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking value counts of new 'loyalty_flag' column

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# 05. Determine whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers

ords_prods_merge.groupby('loyalty_flag').agg({'prices': 'mean'})
```

```python
# 06. Create a spending flag for each user based on the average price across all their orders using the criteria

ords_prods_merge['avg_prod_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check result of previous code

ords_prods_merge.head(15)
```

```python
#creating 'Low spender' flag

ords_prods_merge.loc[ords_prods_merge['avg_prod_price'] < 10, 'spender_type'] = 'Low spender'
```

```python
#creating 'High spender' flag

ords_prods_merge.loc[ords_prods_merge['avg_prod_price'] >= 10, 'spender_type'] = 'High spender'
```

```python
#checking value counts of new flag

ords_prods_merge['spender_type'].value_counts(dropna = False)
```

```python
#checking to see if command ran correctly

ords_prods_merge[['user_id', 'avg_prod_price', 'spender_type']].head(50)
```

```python
# 07. Create an order frequency flag that marks the regularity of a user's ordering behavior according to the median in "days_since_prior_order" column with criteria listed.

#creating "frequency"

ords_prods_merge['frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'days_since_last_order', 'frequency']].head(15)
```

```python
#creating 'Non-freuqent customer' flag

ords_prods_merge.loc[ords_prods_merge['frequency'] > 20, 'frequency_type'] = 'Non-frequent customer'
```

```python
#creating 'Regular customer' flag

ords_prods_merge.loc[(ords_prods_merge['frequency'] > 10) & (ords_prods_merge['frequency'] <= 20), 'frequency_type'] = 'Regular customer'
```

```python
#creating 'Frequent customer' flag

ords_prods_merge.loc[ords_prods_merge['frequency'] <= 10, 'frequency_type'] = 'Frequent customer'
```

```python
#checking 'frequency_type' value count

ords_prods_merge['frequency_type'].value_counts(dropna = False)
```

```python
# Examining the NaN values

print(ords_prods_merge[ords_prods_merge.frequency_type.isnull()])
```

The NaN values for the frequency_type column are attributed by users with only 1 order which results in no median day_since_last_order value; 

```python
#placing NaN value frequency_type flags into "Non-frequent customer" tag

ords_prods_merge['frequency_type'].fillna("Non-frequent customer", inplace = True)
```

```python
#rechecking value counts

ords_prods_merge['frequency_type'].value_counts(dropna = False)
```
