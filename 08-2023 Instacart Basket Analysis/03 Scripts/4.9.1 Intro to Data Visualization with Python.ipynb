```python
#removing NaN values

customers_clean = customers[customers['First Name'].isnull() == False]
```

```python
#checking new dataframe shape

customers_clean.shape
```

```python
#identifying duplicate values

customers_dups = customers_clean[customers_clean.duplicated()]
```

```python
#checking duplicates

customers_dups
```

There are no duplicated values

```python
#Looking to see if there is any data that needs further investigation

customers_clean.describe()
```

```python
#checking for mixed-typed data in dataframe

for col in customers_clean.columns.tolist():
  weird = (customers_clean[[col]].applymap(type) != customers_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers_clean[weird]) > 0:
    print (col)
```

There is no mixed type data
